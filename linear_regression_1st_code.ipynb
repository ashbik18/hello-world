```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#time to load the data
X=[]
Y=[]

for line in open('data.csv'):
    x,y=line.split(',')
    X.append(float(x))
    Y.append(float(y))
X=np.array(X)
Y=np.array(Y)
```

```python
# see how are data looks like
plt.scatter(X,Y)
plt.show()
```

```python
#now performing the gradent decent
m=0
b=0
Learning_rate=0.0001
N=float(len(X))

for i in range (1000):
    Y_hat=m*X+b
    dm=(-2/N)*sum(X*(Y-Y_hat))
    db=(-2/N)*sum(Y-Y_hat)
    
    m=m-Learning_rate*dm
    b=b-Learning_rate*db
    
print('Y ='+str(m)+' X + '+str(b))
plt.scatter(X,Y)
plt.plot(X,Y_hat)
plt.show()
```

```python
#now calculating the R2
d1=Y-Y_hat
d2=Y-Y.mean()
r2=1-(d1.dot(d1)/d2.dot(d2))
print('the value of r2 is given'+str(r2))
```
